```python
# ---------------------------------------------------------------------------------------#
# GBM
# ---------------------------------------------------------------------------------------#

import pandas as pd
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier

training = pd.read_csv("C:/Users/lmrod/OneDrive/Documentos/GitHub/Taller-4-BDML/data/training1.csv")
testing = pd.read_csv("C:/Users/lmrod/OneDrive/Documentos/GitHub/Taller-4-BDML/data/testing1.csv")
test_ori = pd.read_csv("C:/Users/lmrod/OneDrive/Documentos/GitHub/Taller-4-BDML/data/test_ori1.csv")

#X = train_ori.drop('name', axis=1)
#y = train_ori['name']
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_training = training.drop('name', axis=1)
Y_training = training['name']
X_validation = testing.drop('name', axis=1)
Y_validation = testing['name']
X_test = test_ori

from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=3,random_state=0000)
clf = clf.fit(X_training, Y_training)

print('Score, Training: ', clf.score(X_training,Y_training)) #accuracy score on training data = 0.960733384262796
print('Score, Testing: ', clf.score(X_validation,Y_validation)) #accuracy score on validation data 0.7375178316690443

#---------------------------------------Ahora para test

# make a copy of the X_test DataFrame
X_test_copy = X_test.copy()

# drop the column to ignore from the copy
X_test_copy.drop(columns=['id'], inplace=True)

# make predictions on the modified copy
y_pred = clf.predict(X_test_copy)


result = pd.DataFrame({'id': X_test['id'], 'name': y_pred})

# save the result to a file
result.to_csv('C:/Users/lmrod/OneDrive/Documentos/GitHub/Taller-4-BDML/stores/Kaggle_ModeloGB.csv', index=False)

# Accuracy: 0,74


# ---------------------------------------------------------------------------------------#
# RF
# ---------------------------------------------------------------------------------------#

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

#import warnings
#warnings.filterwarnings("ignore")

model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True, verbose=2,
                               max_features = 'sqrt')

model.fit(X_training, Y_training)

y_pred = model.predict(X_test_copy)

result = pd.DataFrame({'id': X_test['id'], 'name': y_pred})

# save the result to a file
result.to_csv('C:/Users/lmrod/OneDrive/Documentos/GitHub/Taller-4-BDML/stores/Kaggle_ModeloRF.csv', index=False)


```
